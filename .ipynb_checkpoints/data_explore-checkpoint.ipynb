```python
import gzip
import json
import pandas as pd
```

```python
path='/Users/mzhang/Desktop/SQL/code/telematics_work_sample/'
```

```python
with gzip.open(path+'mobile_trips.json.gz', "rb") as f:
    mobile = json.loads(f.read().decode("ascii"))
    
with gzip.open(path+'obd2_trips.json.gz', "rb") as f:
    obd2 = json.loads(f.read().decode("ascii"))
```

```python
len(mobile)
```

```python
mobile_table=  pd.DataFrame(columns=['trip_id', 'record_time'], index= range(len(mobile)))
cnt= 0

for i in range(len(mobile)):
    mobile_table.loc[cnt, 'trip_id']= mobile[i][1]['trip_id']
    mobile_table.loc[cnt, 'record_time']= len(mobile[i])
    cnt+=1
```

```python
obd2_table=  pd.DataFrame(columns=['trip_id', 'record_time'], index= range(len(obd2)))
cnt= 0

for i in range(len(obd2)):
    obd2_table.loc[cnt, 'trip_id']= obd2[i][1]['trip_id']
    obd2_table.loc[cnt, 'record_time']= len(obd2[i])
    cnt+=1
```
